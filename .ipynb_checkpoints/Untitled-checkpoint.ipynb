<h1> Data Graphing for Pointcloud Transfer Speeds </h1>

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
x = np.linspace(0,10,20)
y = x**2
plt.plot(x,y)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
fig = plt.figure()
```

```python
ax = fig.add_axes([0.1, 0.2, 0.8, 0.9])
```

```python
ax.plot(x,y, 'purple')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Object Oriented')
```
